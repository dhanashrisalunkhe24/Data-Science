```python
plt.figure(figsize=(10,5))
plt.suptitle('Distribution',fontsize=15)
plt.subplot(1,2,1)
sn.distplot(data4.Waist,label='WAIST')
plt.xlabel('WAIST')
plt.ylabel('Density')
plt.legend()
plt.subplot(1,2,2)
sn.distplot(data4.AT,label='AT')
plt.xlabel('AT')
plt.ylabel('Density')
plt.legend()
```
